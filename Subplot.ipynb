##SUBPLOT
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'bo-')

plt.subplot(1,3,2)
plt.hist(y)

plt.subplot(1,3,3)
plt.plot(y,x,'bo-')

plt.show()
```

ORIENTADO A OBJETOS

```python
x1 = np.linspace(0.1,0.9,10)
y1 = np.sin(x1)

fig, axes = plt.subplots()
axes.plot(x1,y1)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```
