##SUBPLOT
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'bo-')

plt.subplot(1,3,2)
plt.hist(y)

plt.subplot(1,3,3)
plt.plot(y,x,'bo-')

plt.show()
```
