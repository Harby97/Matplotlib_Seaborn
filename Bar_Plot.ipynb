```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Contry = ['INDIA', 'MEXICO', 'COLOMBIA', 'GERMANY', 'CHINA']
Population = [3100, 8300, 3854, 4598, 9999]
```

```python
plt.style.use('seaborn-darkgrid')
plt.bar (Contry, Population, width=0.5, color=['#CD5C5C', '#F08080', '#FA8072', '#E9967A', '#FFA07A'])
plt.xticks(np.arange(5), ('India', 'Mexico', 'Colombia', 'Alemania', 'China'), rotation= 30)
plt.title('Población, paises aleatorios')
plt.xlabel('Paises escogidos')
plt.ylabel('Población en miles')
plt.barh(Contry,Population)
plt.show
```
