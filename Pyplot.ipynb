```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'ro')
```

```python
plt.plot(x,y,'y')
```

```python
plt.plot(x,y,'yx')
```

```python
plt.plot(x,y,'rs-')
```

```python
plt.plot(x,y,'yD:')
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
