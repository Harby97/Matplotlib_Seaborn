```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
sns.countplot(data=tips,x='time',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,color='0',marker='^')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,split=True,palette='pastel')
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.catplot(data=tips,x='time',y='total_bill',hue='sex',palette='pastel',kind='box',col='day')
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.catplot(data=tips,x='day',hue='size',palette='pastel',kind='count');
```
