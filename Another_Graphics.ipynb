```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(0,50,1000)
```

Graficando histogramas, (rangos de frecuencia de los datos)

```python
plt.style.use('seaborn-darkgrid')
plt.hist(data,bins=50, histtype= 'bar',color= '#ffb6af')
plt.show
```

```python
plt.style.use('seaborn-darkgrid')
plt.hist(data,bins=50, histtype= 'step',color= 'black')
plt.show
```

BOX PLOT

```python
data = np.append(data,80)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.show()
```

Scatter

```python
N = 50
x = np.random.random(N)
y = np.random.random(N)
area = (30 * np.random.random(N)) **2
colors = np.random.random(N)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='o',alpha=0.5)
plt.show()
```
