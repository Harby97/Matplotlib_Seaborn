```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.8,1])
axes.plot(x,y,'b')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.8,1])
axes2= fig.add_axes([0.1,0.5,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')
fig.show()
```
