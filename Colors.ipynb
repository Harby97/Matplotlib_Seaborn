```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
plt.style.use('dark_background')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(10,5))
axes[0].plot(x,y,label='$sen(x)$')
axes[0].set_title('Relacion x - y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relacion y - x')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,x+1,'r:',alpha=0.7,linewidth=2,marker='o',markersize=10,markerfacecolor='y')
ax.plot(x,x+2,'g',linewidth=4,marker='x')
ax.plot(x,x+3,'b',linewidth=10,marker='v')
ax.plot(x,x+4,'y',linewidth=1,marker='P')
```
