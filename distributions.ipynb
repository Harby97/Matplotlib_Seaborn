```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.histplot(data=tips, x='tip',bins=20)
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=20,cumulative=True)
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=20,hue='sex')
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=20,hue='sex',stat='frequency')
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=20,hue='sex',stat='probability')
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=20,hue='sex',stat='percent')
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=20,hue='sex',stat='density')
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=20,hue='sex',stat='count',multiple='stack')
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=20,hue='sex',stat='count',multiple='layer')
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=20,hue='sex',stat='count',multiple='dodge')
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=20,hue='sex',stat='count',multiple='fill')
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex')
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',cumulative=False,shade=True,bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='hist',multiple='stack')
plt.show()
plt.savefig('name.png')
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='ecdf')
```

```python
sns.displot(data=tips, x='day', y='time', hue='sex')
```

```python
sns.scatterplot(data=tips,x='tip',y='total_bill',hue='day',size='size',palette='YlGnBu')
```
